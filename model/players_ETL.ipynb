```python
import numpy as np
import pandas as pd
import glob
```

**Clean players' info**

```python
# file path changed to model/file_name.csv 
file_path = "new_matches.csv"

df = pd.read_csv(file_path)
df.head(5)
```

```python
df.columns
```

**winners**

```python
w_df = df[['winner_id','winner_name', 'winner_hand', 'winner_ht'
           , 'winner_ioc', 'winner_age','winner_rank','winner_rank_points']]
w_df
```

```python
w_df2 = w_df.rename(columns={"winner_id": "player_id",
                            "winner_name": "player_name",
                            "winner_hand": "player_hand",
                            "winner_ht": "player_ht",
                            "winner_ioc": "player_ioc",
                            "winner_age": "player_age",
                            "winner_rank" : "player_rank",
                            "winner_rank_points": "player_rank_points"} )
w_df2
```

**loser**

```python
l_df = df[['loser_id', 'loser_name', 'loser_hand','loser_ht', 'loser_ioc', 'loser_age','loser_rank','loser_rank_points']]
l_df
```

```python
l_df2 = l_df.rename(columns={"loser_id": "player_id",
                            "loser_name": "player_name",
                            "loser_hand": "player_hand",
                            "loser_ht": "player_ht",
                            "loser_ioc": "player_ioc",
                            "loser_age": "player_age",
                            "loser_rank":"player_rank",
                            "loser_rank_points":"player_rank_points"})
l_df2
```

```python
all_players = pd.concat([w_df2, l_df2], axis = 0)
all_players
```

```python
all_players['player_age'] = round(all_players['player_age'])
all_players
```

```python
all_players.isnull().sum()
```

```python
all_players.duplicated()
```

```python
all_players.shape
```

```python
all_players1 = all_players.sort_values('player_age', ascending=False).drop_duplicates('player_id').sort_index()
all_players1
```

```python
all_players1.shape
```

```python
all_players1.fillna('Information Not Available', inplace=True)
all_players1.sample(10)
```

```python
all_players1.sort_values('player_name', ascending=True, inplace=True)
all_players1.sample(10)
```

### new_rankings

```python
file_path = "new_rankings.csv"

rankings_df = pd.read_csv(file_path)
rankings_df.head(5)
```

```python
rankings = rankings_df.rename(columns={"player": "player_id"})
rankings
```

### Merge players and rankings: 
`all_players1` and `rankings` on `player_id`
We want Player Names on Rankings info

```python
all_player2 = all_players[['player_id','player_name']]
all_player2 =  all_player2.drop_duplicates()
all_player2.index = range(all_player2.shape[0])
all_player2
```

```python
pen = pd.merge(rankings,all_player2, left_on="player_id",right_on='player_id')
```

```python
pen.sort_values(by='ranking_date')
```
