```python
import numpy as np
import pandas as pd
import glob
```

**Clean players' info**

```python
file_path = "Resources/new_matches.csv"

df = pd.read_csv(file_path)
df.head(5)
```

```python
df.columns
```

**winners**

```python
w_df = df[['winner_id','winner_name', 'winner_hand', 'winner_ht', 'winner_ioc', 'winner_age']]
w_df
```

```python
w_df2 = w_df.rename(columns={"winner_id": "player_id",
                            "winner_name": "player_name",
                            "winner_hand": "player_hand",
                            "winner_ht": "player_ht",
                            "winner_ioc": "player_ioc",
                            "winner_age": "player_age"})
w_df2
```

**loser**

```python
l_df = df[['loser_id', 'loser_name', 'loser_hand','loser_ht', 'loser_ioc', 'loser_age']]
l_df
```

```python
l_df2 = l_df.rename(columns={"loser_id": "player_id",
                            "loser_name": "player_name",
                            "loser_hand": "player_hand",
                            "loser_ht": "player_ht",
                            "loser_ioc": "player_ioc",
                            "loser_age": "player_age"})
l_df2
```

```python
all_players = pd.concat([w_df2, l_df2], axis = 0)
all_players
```

```python
all_players['player_age'] = round(all_players['player_age'])
all_players
```

```python
all_players.isnull().sum()
```

```python
all_players.duplicated()
```

```python
all_players.shape
```

```python
all_players.sort_values("player_id", inplace = True)
all_players
```

```python
new_all_players = all_players.drop_duplicates(subset=['player_id','player_name','player_age'], keep='last')
new_all_players
```
